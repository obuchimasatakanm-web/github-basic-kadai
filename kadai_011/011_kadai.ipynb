リストの中身を先頭から全て取り出して、出力結果が以下のようになるプログラムを、for、while、それぞれを利用して記述して下さい。

※forループで水～冥の9行を表示し、つぎにwhileループで水～冥の9行を表示してください

```python
array = ["水","金","地","火","木","土","天","海","冥"]
i = 0
count = len(array)

# for文
for str in array:
    print(str)

#while文　要素の数になるまで繰り返す。
while i <= count:
    if i == count:
       break
    else:
       print(array[i])
       i += 1
       continue
```
