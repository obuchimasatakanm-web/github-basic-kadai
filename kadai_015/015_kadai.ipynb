名前(name)と年齢(age)の属性を持つHumanクラスを作成してください。
nameとageを標準出力(print)するメソッド(printinfo)を追加してください。

Humanクラスのインスタンスは、変数に代入してプログラム内で使用してください。

```python
class Human:
    #コンストラクタ定義
    def __init__(self, name, age):
        self.name = name
        self.age = age

    #メソッド呼び出し
    def printinfo(self):
        print(self.name)
        print(self.age)

#インスタンス化
human = Human("太郎", 20)

# メソッド呼び出し
human.printinfo()

```
