名前(name)と年齢(age)の属性を持つHumanクラスを作成してください。

Humanクラスには、以下の条件で標準出力(print)するcheck_adultメソッドを追加してください。

ageが20以上の場合に大人であること

そうでない場合に大人でないこと

Humanクラスのインスタンスを複数生成してリストに追加し、リストの要素数分だけcheck_adultメソッドを呼び出してください。

```python
# 年齢リスト
list = []

class Human:
    #コンストラクタ定義
    def __init__(self, name, age):
        self.name = name
        self.age = age

    #メソッド呼び出し
    def check_adult(self, age):
        if age >= 20:
            print("大人です")
        else:
            print("大人ではありません")

    #リスト追加
    def add_age(self):
        #print(self.name)
        #print(self.age)
        list.append(self.age)

#インスタンス化
human = Human("太郎", 20)
human.add_age()
human = Human("次郎", 18)
human.add_age()
human = Human("三郎", 16)
human.add_age()
human = Human("花子", 30)
human.add_age()

# for文でリストの要素数分だけ判定
i = 0
count = len(list)

for str in list:
    human.check_adult(str)
```
