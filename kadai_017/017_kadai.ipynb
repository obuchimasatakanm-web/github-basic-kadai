名前(name)と年齢(age)の属性を持つHumanクラスを作成してください。

Humanクラスには、以下の条件で標準出力(print)するcheck_adultメソッドを追加してください。

ageが20以上の場合に大人であること

そうでない場合に大人でないこと

Humanクラスのインスタンスを複数生成してリストに追加し、リストの要素数分だけcheck_adultメソッドを呼び出してください。

```python


class Human:
    #コンストラクタ定義
    def __init__(self, name, age):
        self.name = name
        self.age = age

    #メソッド呼び出し
    def check_adult(self):
        if self.age >= 20:
            return "大人です"
        else:
            return "大人ではありません"

    #リスト追加
    #def print_info(self):
        # print(self.name, self.age)

#インスタンス化
#年齢判定リスト
humans = []
human = Human("太郎", 20)
humans.append(human)
human = Human("次郎", 18)
humans.append(human)
human = Human("三郎", 16)
humans.append(human)
human = Human("花子", 30)
humans.append(human)

# # for文で判定
for str in humans:
    result = str.name + "は、" + str.check_adult()
    print(result)
```
